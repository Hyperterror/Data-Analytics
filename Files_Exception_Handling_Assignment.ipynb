## Practical 1. Open a file for writing

```python
with open("output.txt", "w") as f:
    f.write("Hello World")
```

## Practical 2. Read a file and print each line

```python
with open("output.txt", "r") as f:
    for line in f:
        print(line.strip())
```

## Practical 3. Handle missing file

```python
try:
    with open("missing.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found")
```

## Practical 4. Copy one file to another

```python
with open("output.txt", "r") as src, open("copy.txt", "w") as dst:
    dst.write(src.read())
```

## Practical 5. Handle division by zero

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

## Practical 6. Log error message on division by zero

```python
import logging
logging.basicConfig(filename="error.log", level=logging.ERROR)
try:
    print(10 / 0)
except ZeroDivisionError as e:
    logging.error("Division by zero occurred: %s", e)
```

## Practical 7. Log at INFO, ERROR, WARNING levels

```python
import logging
logging.basicConfig(level=logging.INFO)
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
```

## Practical 8. File opening error handling

```python
try:
    f = open("abc.txt", "r")
except Exception as e:
    print("Error:", e)
```

## Practical 9. Read file into list

```python
with open("output.txt", "r") as f:
    lines = f.readlines()
print(lines)
```

## Practical 10. Append data to file

```python
with open("output.txt", "a") as f:
    f.write("\nAppended text")
```

## Practical 11. Handle missing dictionary key

```python
d = {"a": 1}
try:
    print(d["b"])
except KeyError:
    print("Key not found")
```

## Practical 12. Multiple except blocks

```python
try:
    x = int("abc")
except ValueError:
    print("Value error")
except TypeError:
    print("Type error")
```
