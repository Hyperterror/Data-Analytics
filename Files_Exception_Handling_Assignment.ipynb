## Theory Answers

**1. What is the difference between interpreted and compiled languages?**
- *Compiled* languages (C, C++) translate the entire source code to machine code before execution, producing a standalone executable. This is generally faster at runtime.
- *Interpreted* languages (Python, JavaScript) execute code line by line at runtime through an interpreter. This makes them more portable and easier to debug but slightly slower.

**2. What is exception handling in Python?**
Exception handling is a mechanism to detect and respond to runtime errors (exceptions) without crashing the program. Python uses `try`, `except`, `else`, and `finally` blocks. When an error occurs inside a `try` block, Python looks for a matching `except` block to handle it gracefully.

**3. What is the purpose of the `finally` block in exception handling?**
The `finally` block always executes regardless of whether an exception occurred or not. It is used for cleanup tasks that must always run, such as closing file handles, releasing locks, or disconnecting from a database.

**4. What is logging in Python?**
Logging is the process of recording runtime events to a log file or console using Python's built-in `logging` module. It supports severity levels: DEBUG, INFO, WARNING, ERROR, and CRITICAL. Unlike `print()`, logs can be configured, filtered, formatted, and written to files for long-term auditing.

**5. What is the significance of the `__del__` method in Python?**
`__del__` is the destructor method, called by Python's garbage collector just before an object is destroyed. It can be used to release external resources (e.g., network connections, file handles) when an object is no longer needed. However, the exact timing of `__del__` is not guaranteed, so `with` statements are generally preferred.

**6. What is the difference between `import` and `from ... import` in Python?**
- `import module` imports the entire module; you access names as `module.name`.
- `from module import name` imports a specific name directly into the current namespace; you access it as `name` without the module prefix.
- `from module import *` imports all public names (not recommended due to namespace pollution).

**7. How can you handle multiple exceptions in Python?**
You can handle multiple exceptions by:
1. Using separate `except` clauses for each exception type.
2. Grouping them in a single `except` clause using a tuple: `except (ValueError, TypeError):`.
3. Using a bare `except:` or `except Exception:` to catch any exception (not recommended for broad use).

**8. What is the purpose of the `with` statement when handling files in Python?**
The `with` statement (context manager) ensures that a resource such as a file is automatically opened and properly closed after the block executes, even if an exception occurs. It replaces manual `try/finally` for resource management and makes code cleaner and safer.

**9. What is the difference between multithreading and multiprocessing?**
- *Multithreading*: Multiple threads run within the same process and share memory. Python's GIL (Global Interpreter Lock) limits true CPU parallelism in threads; best for I/O-bound tasks.
- *Multiprocessing*: Multiple processes run with separate memory spaces, bypassing the GIL. Best for CPU-bound tasks that need true parallel execution.

**10. What are the advantages of using logging in a program?**
- Configurable severity levels allow selective output.
- Logs can be written to files, streams, or external systems.
- Log messages include timestamps, module names, and line numbers automatically.
- Logging is thread-safe and suitable for production systems.
- Unlike `print()`, logging can be enabled or disabled without code changes.

**11. What is memory management in Python?**
Python manages memory automatically through:
- *Reference counting*: Each object tracks how many references point to it; when count reaches 0, memory is freed.
- *Garbage collector*: Handles cyclic references that reference counting cannot resolve, using a generational GC.
- *Memory pools*: Python maintains private heap space managed by the `pymalloc` allocator.

**12. What are the basic steps involved in exception handling in Python?**
1. `try`: Place the code that might raise an exception.
2. `except`: Catch and handle the specific exception.
3. `else`: Executed if no exception was raised in the `try` block.
4. `finally`: Always executed; used for cleanup.

**13. Why is memory management important in Python?**
Proper memory management prevents memory leaks (where objects are never freed), reduces unnecessary memory usage, and improves application performance and stability. In production applications, unmanaged memory can exhaust system resources and cause crashes.

**14. What is the difference between `read()`, `readline()`, and `readlines()` in Python file handling?**
- `read()`: Reads the entire file as a single string.
- `readline()`: Reads one line at a time; each call returns the next line.
- `readlines()`: Reads all lines and returns them as a list of strings, one element per line.

## Practical 1. Open a file for writing

```python
with open("output.txt", "w") as f:
    f.write("Hello World")
```

## Practical 2. Read a file and print each line

```python
with open("output.txt", "r") as f:
    for line in f:
        print(line.strip())
```

## Practical 3. Handle missing file

```python
try:
    with open("missing.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found")
```

## Practical 4. Copy one file to another

```python
with open("output.txt", "r") as src, open("copy.txt", "w") as dst:
    dst.write(src.read())
```

## Practical 5. Handle division by zero

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

## Practical 6. Log error message on division by zero

```python
import logging
logging.basicConfig(filename="error.log", level=logging.ERROR)
try:
    print(10 / 0)
except ZeroDivisionError as e:
    logging.error("Division by zero occurred: %s", e)
```

## Practical 7. Log at INFO, ERROR, WARNING levels

```python
import logging
logging.basicConfig(level=logging.INFO)
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
```

## Practical 8. File opening error handling

```python
try:
    f = open("abc.txt", "r")
except Exception as e:
    print("Error:", e)
```
