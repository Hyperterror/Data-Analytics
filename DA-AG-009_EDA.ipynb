## Q1. Read the Bike Details dataset into a Pandas DataFrame and display its first 10 rows. Show the shape and column names.

```python
df = pd.read_csv('bike_details.csv')\nprint('Shape:', df.shape)\nprint('Columns:', df.columns.tolist())\ndf.head(10)
```

## Q2. Check for missing values in all columns and describe your approach for handling them.

```python
print(df.isna().sum())\n\n# Approach: No missing values here. Generally: drop rows/cols >50% missing, impute median for numeric, mode for categorical.
```

## Q3. Plot the distribution of selling prices using a histogram and describe the overall trend.

```python
plt.figure(figsize=(8,5))\nplt.hist(df['selling_price'], bins=5, edgecolor='black')\nplt.title('Selling Price Distribution')\nplt.xlabel('Selling Price')\nplt.ylabel('Frequency')\nplt.show()\n\n# Trend: Prices mostly between 20k-40k, right-skewed towards higher end.
```

## Q4. Create a bar plot to visualize the average selling price for each seller type and write one observation.

```python
avg_price = df.groupby('seller_type')['selling_price'].mean()\navg_price.plot(kind='bar')\nplt.title('Avg Selling Price by Seller Type')\nplt.ylabel('Avg Price')\nplt.xticks(rotation=0)\nplt.show()\n\n# Observation: Dealers have higher average prices than Individuals.
```

## Q5. Compute the average km driven for each ownership type and present the result as a bar plot.

```python
avg_km = df.groupby('owner')['km_driven'].mean()\navg_km.plot(kind='bar')\nplt.title('Avg KM Driven by Owner Type')\nplt.ylabel('Avg KM')\nplt.xticks(rotation=0)\nplt.show()
```

## Q6. Use the IQR method to detect and remove outliers from the km driven column. Show before-and-after summary statistics.

```python
Q1 = df['km_driven'].quantile(0.25)\nQ3 = df['km_driven'].quantile(0.75)\nIQR = Q3 - Q1\nlower = Q1 - 1.5*IQR\nupper = Q3 + 1.5*IQR\n\nprint('Before:\n', df['km_driven'].describe())\ndf_no_outliers = df[(df['km_driven'] >= lower) & (df['km_driven'] <= upper)]\nprint('\nAfter:\n', df_no_outliers['km_driven'].describe())
```

## Q7. Create a scatter plot of year vs. selling price to explore the relationship between bike age and price.

```python
plt.figure(figsize=(8,5))\nplt.scatter(df['year'], df['selling_price'], alpha=0.7)\nplt.title('Year vs Selling Price')\nplt.xlabel('Year')\nplt.ylabel('Selling Price')\nplt.show()\n# Newer years correlate with higher prices.
```

## Q8. Convert the seller type column into numeric format using one-hot encoding. Display the first 5 rows.

```python
df_encoded = pd.get_dummies(df, columns=['seller_type'], prefix='seller')\ndf_encoded.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])\nplt.figure(figsize=(8,6))\nsns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')\nplt.title('Correlation Heatmap')\nplt.show()\n# Strongest: Year vs Price (newer = pricier), negative KM vs Price.
```
