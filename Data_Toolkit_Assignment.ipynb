## Practical 1. 2D NumPy array and row sums

```python
import numpy as np
a = np.array([[1,2,3],[4,5,6]])
print(a.sum(axis=1))
```

## Practical 2. Mean of a specific DataFrame column

```python
import pandas as pd
df = pd.DataFrame({"A":[10,20,30],"B":[1,2,3]})
print(df["A"].mean())
```

## Practical 3. Scatter plot using Matplotlib

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.scatter(x,y)
plt.show()
```

## Practical 4. Correlation matrix with heatmap

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
df = pd.DataFrame({"A":[1,2,3],"B":[2,4,6],"C":[5,6,7]})
sns.heatmap(df.corr(), annot=True)
plt.show()
```

## Practical 5. Bar plot using Plotly

```python
import plotly.express as px
df = px.data.tips()
fig = px.bar(df, x="day", y="total_bill")
fig.show()
```

## Practical 6. Add a new DataFrame column

```python
import pandas as pd
df = pd.DataFrame({"marks":[50,60,70]})
df["double_marks"] = df["marks"] * 2
print(df)
```

## Practical 7. Element-wise multiplication of arrays

```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a * b)
```

## Practical 8. Multiple line plot

```python
import matplotlib.pyplot as plt
x = [1,2,3]
y1 = [1,4,9]
y2 = [1,2,3]
plt.plot(x,y1,label="square")
plt.plot(x,y2,label="linear")
plt.legend()
plt.show()
```

## Practical 9. Filter rows in DataFrame

```python
import pandas as pd
df = pd.DataFrame({"A":[10,25,30]})
print(df[df["A"] > 20])
```
