## Q1. What is Simple Linear Regression?

Simple linear regression models the relationship between one independent variable and one dependent variable using a straight line.

## Q2. What are the key assumptions of Simple Linear Regression?

The main assumptions are linearity, independence, homoscedasticity, normality of residuals, and minimal multicollinearity when extended to multiple predictors.

## Q3. Write the mathematical equation for a simple linear regression model and explain each term.

The equation is y = β0 + β1x + ε, where y is the dependent variable, β0 is the intercept, β1 is the slope, x is the independent variable, and ε is the error term.

## Q4. Provide a real-world example where simple linear regression can be applied.

It can be used to predict house price based on area or salary based on years of experience.

## Q5. What is the method of least squares in linear regression?

The least squares method chooses the line that minimizes the sum of squared differences between observed and predicted values.

## Q6. What is Logistic Regression? How does it differ from Linear Regression?

Logistic regression is used for classification problems and predicts probabilities, while linear regression predicts continuous values.

## Q7. Name and briefly describe three common evaluation metrics for regression models.

Common regression metrics are MAE, MSE, and RMSE. MAE measures average absolute error, MSE penalizes larger errors more, and RMSE is the square root of MSE and is in the same unit as the target.

## Q8. What is the purpose of the R-squared metric in regression analysis?

R-squared measures the proportion of variance in the dependent variable explained by the model.

## Q9. Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,4,5,4,5])
model = LinearRegression().fit(X, y)
print('Slope:', round(model.coef_[0], 3))
print('Intercept:', round(model.intercept_, 3))
```

## Q10. How do you interpret the coefficients in a simple linear regression model?

The intercept is the predicted value when the independent variable is zero, and the slope tells how much the dependent variable changes for each one-unit increase in the independent variable.
