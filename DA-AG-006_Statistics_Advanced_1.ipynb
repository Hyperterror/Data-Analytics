## Q1. What is a random variable in probability theory?

A random variable is a numerical representation of the outcome of a random process. It assigns a number to each possible outcome so that probabilistic analysis can be performed.

## Q2. What are the types of random variables?

The two main types are discrete random variables, which take countable values, and continuous random variables, which take any value within an interval.

## Q3. Explain the difference between discrete and continuous distributions.

Discrete distributions describe countable outcomes such as the number of heads in coin flips, while continuous distributions describe measurements such as height or time that can take infinitely many values within a range.

## Q4. What is a binomial distribution, and how is it used in probability?

A binomial distribution models the number of successes in a fixed number of independent trials, where each trial has only two outcomes and the success probability remains constant. It is used for events like pass/fail, yes/no, or success/failure experiments.

## Q5. What is the standard normal distribution, and why is it important?

The standard normal distribution is a normal distribution with mean 0 and standard deviation 1. It is important because many statistical methods use z-scores and probabilities derived from this reference distribution.

## Q6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, even if the population is not normal. It is critical because it justifies confidence intervals and many hypothesis tests.

## Q7. What is the significance of confidence intervals in statistical analysis?

A confidence interval gives a plausible range for a population parameter based on sample data. It helps quantify uncertainty and shows the precision of an estimate.

## Q8. What is the concept of expected value in a probability distribution?

Expected value is the long-run average outcome of a random variable, computed as a weighted average of all possible values using their probabilities.

## Q9. Python program to generate 1000 random numbers from a normal distribution with mean 50 and standard deviation 5.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)
mean = data.mean()
std = data.std()
print('Mean:', round(mean, 2))
print('Std Dev:', round(std, 2))

plt.hist(data, bins=30, edgecolor='black')
plt.title('Normal Distribution Sample')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Q10. Apply CLT to estimate average sales with a 95% confidence interval.

The CLT lets us estimate the average sales using the sample mean and a confidence interval. For the given sales data, the mean is about 248.25 and the 95% confidence interval is approximately 240.47 to 256.03.

```python
import numpy as np
from scipy import stats

sales = np.array([220,245,210,265,230,250,260,275,240,255,235,260,245,250,225,270,265,255,250,260])
mean = sales.mean()
sem = stats.sem(sales)
ci = stats.t.interval(0.95, len(sales)-1, loc=mean, scale=sem)
print('Mean:', round(mean, 2))
print('95% CI:', tuple(round(x, 2) for x in ci))
```
