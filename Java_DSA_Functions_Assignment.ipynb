## Practical 1. Sum of even numbers

```python
def sum_even(nums):
    return sum(x for x in nums if x % 2 == 0)
```

## Practical 2. Reverse string

```python
def reverse_string(s):
    return s[::-1]
```

## Practical 3. Square each integer

```python
def square_list(nums):
    return [x**2 for x in nums]
```

## Practical 4. Check prime from 1 to 200

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True
```

## Practical 5. Fibonacci iterator class

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b, self.count = 0, 1, 0
    def __iter__(self): return self
    def __next__(self):
        if self.count >= self.n: raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val
```

## Practical 6. Generator powers of 2

```python
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i
```

## Practical 7. Read file line by line generator

```python
def read_lines(file_path):
    with open(file_path, "r") as f:
        for line in f:
            yield line.strip()
```

## Practical 8. Sort tuples by second element

```python
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

## Practical 9. Celsius to Fahrenheit using map

```python
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

## Practical 10. Remove vowels using filter

```python
text = "Hello World"
result = "".join(filter(lambda ch: ch.lower() not in "aeiou", text))
print(result)
```

## Practical 11. Book shop accounting using lambda and map

```python
orders = [["34587", "Learning Python", 4, 40.95], ["98762", "Data Science", 1, 98.90], ["77226", "Machine Learning", 3, 24.99]]
result = list(map(lambda x: (x[0], x[2] * x[3] + 10 if x[2] * x[3] < 100 else x[2] * x[3]), orders))
print(result)
```
