## Q1. Explain the differences between AI, ML, Deep Learning, and Data Science.

AI is the broad field of making machines intelligent. ML is a subset of AI that learns patterns from data. Deep Learning is a subset of ML that uses neural networks with multiple layers. Data Science focuses on extracting insights from data using statistics, programming, and domain knowledge.

## Q2. What are the types of machine learning? Describe each with one real-world example.

The main types are supervised learning, unsupervised learning, semi-supervised learning, and reinforcement learning. Supervised learning uses labeled data, unsupervised learning finds structure in unlabeled data, semi-supervised uses a mix, and reinforcement learning learns by rewards and penalties.

## Q3. Define overfitting, underfitting, and the bias-variance tradeoff.

Overfitting happens when a model learns noise and performs well on training data but poorly on new data. Underfitting happens when a model is too simple to capture patterns. The bias-variance tradeoff balances simplicity and flexibility so the model generalizes well.

## Q4. What are outliers in a dataset, and list three common techniques for handling them.

Outliers are extreme values that differ strongly from the rest of the data. Common handling methods include removing them, capping them using winsorization, and transforming the data using log or other transformations.

## Q5. Explain handling missing values and mention one imputation technique for numerical and categorical data.

Missing values can be handled by deletion, simple imputation, or model-based methods. For numerical data, mean or median imputation is common. For categorical data, mode imputation is commonly used.

## Q7. Implement one-hot encoding using pandas for Red, Green, Blue, Green, Red.

## Q8. Generate 1000 samples from a normal distribution, introduce 50 missing values, fill with mean, and plot before/after.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
arr = np.random.normal(0, 1, 1000)
idx = np.random.choice(1000, 50, replace=False)
arr[idx] = np.nan
df = pd.DataFrame({'x': arr})
filled = df['x'].fillna(df['x'].mean())
fig, ax = plt.subplots(1,2,figsize=(10,4))
ax[0].hist(df['x'].dropna(), bins=30, edgecolor='black')
ax[0].set_title('Before')
ax[1].hist(filled, bins=30, edgecolor='black')
ax[1].set_title('After')
plt.show()
```
