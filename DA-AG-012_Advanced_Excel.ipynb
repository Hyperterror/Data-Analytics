## Q1. Explain the difference between Absolute, Relative, and Mixed Cell Referencing in Excel with examples.

Relative references change when copied, absolute references stay fixed with dollar signs, and mixed references lock either the row or the column.

## Q2. What is a Macro in Excel? How does it help in automation?

A macro is a recorded or written set of actions in VBA that automates repetitive Excel tasks and saves time.

## Q3. What are Text Functions in Excel? Mention any five with examples.

Text functions manipulate strings. Common examples include LEFT, RIGHT, MID, LEN, and CONCAT or TEXTJOIN.

## Q4. What is the use of Scenario Manager in decision making?

Scenario Manager lets users compare different input combinations and see how they affect the final result, which helps in what-if analysis.

## Q5. Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best?

VLOOKUP searches vertically and HLOOKUP searches horizontally. XLOOKUP is more flexible because it can search in any direction and is easier to use. In modern Excel, XLOOKUP is usually the best option.
