## Q1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It evaluates whether observed evidence is strong enough to reject a null hypothesis.

## Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis states that there is no effect or no difference. The alternative hypothesis states that an effect or difference exists.

## Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

The significance level, usually denoted by alpha, is the threshold for rejecting the null hypothesis. If the p-value is less than alpha, the null hypothesis is rejected.

## Q4. What are Type I and Type II errors? Give examples of each.

A Type I error occurs when a true null hypothesis is rejected, such as falsely claiming a medicine works. A Type II error occurs when a false null hypothesis is not rejected, such as missing a real treatment effect.

## Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

A Z-test is used when population variance is known or sample size is large, while a T-test is used when population variance is unknown and sample size is small.

## Q6. Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.binomial(n=10, p=0.5, size=1000)
print('Mean:', round(data.mean(), 2))
print('Std Dev:', round(data.std(), 2))
plt.hist(data, bins=range(12), edgecolor='black', align='left')
plt.title('Binomial Distribution')
plt.xlabel('Successes')
plt.ylabel('Frequency')
plt.show()
```

## Q7. Implement hypothesis testing using Z-statistics for the sample dataset in Python.

Assuming the population mean is 50 and population standard deviation is known, the sample mean is very close to 50, so the Z-statistic is near 0 and the result does not provide evidence to reject the null hypothesis.

```python
import numpy as np
from scipy import stats

sample = np.array([49.1,50.2,51.0,48.7,50.5,49.8,50.3,50.7,50.2,49.6,50.1,49.9,50.8,50.4,48.9,50.6,50.0,49.7,50.2,49.5,50.1,50.3,50.4,50.5,50.0,50.7,49.3,49.8,50.2,50.9,50.3,50.4,50.0,49.7,50.5,49.9])
mu0 = 50
sigma = 0.5
z = (sample.mean() - mu0) / (sigma / np.sqrt(len(sample)))
p = 2 * (1 - stats.norm.cdf(abs(z)))
print('Z:', round(z, 3))
print('p-value:', round(p, 4))
```

## Q8. Simulate normal data and calculate the 95% confidence interval for its mean.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(100, 15, 500)
mean = data.mean()
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=stats.sem(data))
print('Mean:', round(mean, 2))
print('95% CI:', tuple(round(x, 2) for x in ci))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Simulated Normal Data')
plt.show()
```

## Q9. Calculate Z-scores from a dataset and visualize the standardized data using a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

data = np.array([10, 12, 14, 16, 18, 20, 22])
z = zscore(data)
print(np.round(z, 3))
plt.hist(z, bins=10, edgecolor='black')
plt.title('Z-score Distribution')
plt.show()
```
