## Q1. Descriptive vs Inferential Statistics

Descriptive statistics summarize and describe data, while inferential statistics draw conclusions about a population from a sample.

## Q2. Sampling

Sampling means selecting a subset of a population. Random sampling gives equal chance to each member, while stratified sampling divides the population into groups before sampling.

## Q3. Mean, Median, Mode

Mean is the average, median is the middle value, and mode is the most frequent value. They are important because they describe the center of the data.

## Q4. Skewness and Kurtosis

Skewness measures asymmetry and kurtosis measures how heavy the tails are. Positive skew means the distribution has a longer tail on the right side.

## Q5. Practical: Mean, Median, Mode

```python
import numpy as np
from statistics import mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
print("Mean:", np.mean(numbers))
print("Median:", np.median(numbers))
print("Mode:", mode(numbers))
```

## Q6. Practical: Covariance and Correlation

```python
import numpy as np

listx = [10, 20, 30, 40, 50]
listy = [15, 25, 35, 45, 60]
print("Covariance:
", np.cov(listx, listy))
print("Correlation:", np.corrcoef(listx, listy)[0,1])
```

## Q7. Practical: Boxplot and Outliers

```python
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```
The boxplot helps detect outliers visually. In this list, 35 may appear as a high-end outlier depending on the IQR boundary.

## Q8. Practical: Advertising Spend and Sales Correlation

```python
import numpy as np

ad_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
print("Correlation:", np.corrcoef(ad_spend, daily_sales)[0,1])
```
A strong positive correlation means sales tend to increase as advertising spend increases.

## Q9. Practical: Histogram for Survey Scores

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
plt.hist(survey_scores, bins=6, edgecolor="black")
plt.title("Customer Satisfaction Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
Useful summary statistics include mean, median, standard deviation, and histogram-based distribution shape.

## Q10. Summary

For survey data, use mean, median, standard deviation, minimum, maximum, and a histogram to understand central tendency, spread, and shape of the distribution.
