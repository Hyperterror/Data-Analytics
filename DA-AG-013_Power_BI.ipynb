## Q1. Define Power BI and its key components.

Power BI is a business intelligence platform for transforming data into interactive reports and dashboards. Its main components include Power BI Desktop, Power BI Service, Power BI Mobile, and Power BI Gateway.

## Q2. Compare Pie Chart vs Donut Chart and Bar Chart vs Column Chart.

Pie and donut charts both show parts of a whole, but donut charts leave space in the center for labels or metrics. Bar charts are horizontal and work well for long category names, while column charts are vertical and useful for time series or ordered categories.

## Q3. Explain Star schema vs Snowflake schema, primary key vs foreign key, and cardinality.

A star schema keeps dimensions denormalized around a fact table, while snowflake schema normalizes dimensions further. Primary keys uniquely identify rows, foreign keys link tables, and cardinality defines how records relate across tables.

## Q4. Differentiate calculated column vs measure. Define row context and filter context.

Calculated columns are computed row by row and stored in the model, while measures are calculated on demand. Row context evaluates one row at a time, and filter context depends on the filters applied in visuals or slicers.

## Q5. What is the difference between a report and a dashboard in Power BI?

A report usually contains multiple pages and detailed interactive visuals, while a dashboard is a single-page summary built from pinned visuals.
